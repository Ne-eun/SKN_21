```python
import pandas as pd
train_df = pd.read_csv('data/original_train.csv')
train_df.shape
```

```python
matchs = pd.read_csv('data/match_info.csv')
matchs.head()
```

전체 데이터는 2024년 데이터.  
10월 18일 이후 11월 데이터 -> test  
9월 20일 이후 부터 10월 18일 이전까지 데이터 -> valid  
그외 모든 데이터 -> train


```python
valid_matchs = matchs[matchs['game_date'] > '2024-09-20']
valid_match_ids = valid_matchs.game_id.unique()
valid_match_ids
```

```python
valid_df = train_df[train_df.game_id.isin(valid_match_ids)]
valid_df.to_csv('data/valid.csv', index=False)
```

```python
train_df = train_df[~train_df.game_id.isin(valid_match_ids)]
train_df.to_csv('data/train.csv', index=False)
```
