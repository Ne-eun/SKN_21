```python
import pandas as pd
import matplotlib.pyplot as plt
from modules.get_data_table import get_base_data
data = get_base_data()
```

```python
# 무게도없고 수량도없는데 무역량만 있는 hs4 코드 확인
data.query("(weight < 1) & (quantity < 1) & (value > 0)").hs4.unique()
```

```python
# hs4 코드별 연관성이 있는지 확인
df = data.groupby("hs4")[['weight', 'quantity', 'value']].agg(['min', 'max', 'sum'])
df[df[('value', 'min')] < 1]
```

```python
data[(data['hs4'] == 2710) & (data['quantity'] > 0)]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Weight statistics
weight_stats = df['weight']
axes[0, 0].scatter(weight_stats['sum'], weight_stats['max'])
axes[0, 0].set_xlabel('Weight Sum')
axes[0, 0].set_ylabel('Weight Max')
axes[0, 0].set_title('Weight: Sum vs Max')

# Quantity statistics
quantity_stats = df['quantity']
axes[0, 1].scatter(quantity_stats['sum'], quantity_stats['max'])
axes[0, 1].set_xlabel('Quantity Sum')
axes[0, 1].set_ylabel('Quantity Max')
axes[0, 1].set_title('Quantity: Sum vs Max')

# Value statistics
value_stats = df['value']
axes[1, 0].scatter(value_stats['sum'], value_stats['max'])
axes[1, 0].set_xlabel('Value Sum')
axes[1, 0].set_ylabel('Value Max')
axes[1, 0].set_title('Value: Sum vs Max')

# Combined comparison
axes[1, 1].scatter(df['weight']['sum'], df['value']['sum'])
axes[1, 1].set_xlabel('Weight Sum')
axes[1, 1].set_ylabel('Value Sum')
axes[1, 1].set_title('Weight Sum vs Value Sum')

plt.tight_layout()
plt.show()
```
